```python
import networkx as nx 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
g = nx.erdos_renyi_graph(100,.1)
nx.draw(g,with_labels=True)
plt.show()
```

```python
def plot_fun():
	node_cnt = g.number_of_nodes()
	max_degree = 0
	for i in g.nodes():
		max_degree = max(max_degree,i)
	x=[]
	y=[]
	for i in range(max_degree+1):
		x.append(i)
		y.append(0)
		for j in g.nodes():
			if(g.degree(j)==i):
				y[i]+=1
		y1 = [i/node_cnt for i in y]
	plt.plot(x,y1,'o')
	plt.xscale('log')
	plt.yscale('log')
	plt.title('Degree Distribution(log-log scale)')
	y2 = [i for i in range(1,10)]
	x1 = []
	for i in y2:
		j = (i**-3)
		x1.append(j)
	plt.plot(y2,x1)
	plt.xlabel('k')
	plt.ylabel('P(k')
	plt.show()
	plt.hist(x1)
	slope,intercept = np.polyfit(y2,x1,1)
	print(slope)
	plt.show()
```

```python
plot_fun()
```
